### Packages &#x1F4DA;

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv('ase_completa.csv', sep=';')
```

### Figure 6.1 &#x1F308;	

```python
correlation_matrix = df.drop(['frameTime','file'], axis=1).corr()

target = 'Arousal'

top_10_features = correlation_matrix[target].abs().sort_values(ascending=False).head(11).index 

corr_top10 = correlation_matrix.loc[top_10_features, top_10_features]

plt.figure(figsize=(10, 8))
sns.heatmap(corr_top10, annot=True, cmap='vlag', fmt='.2f', linewidths=0.5)
plt.title(f'Top 10 Most Correlated Features with {target}')
#plt.savefig('A_most-correlated-features.pdf', bbox_inches='tight')
plt.show()
```

### Figure 6.2 &#x1F308;	

```python
correlation_matrix = df.drop(['frameTime','file'], axis=1).corr()

target = 'Valence'

top_10_features = correlation_matrix[target].abs().sort_values(ascending=False).head(11).index  # Inclui o prprio target

corr_top10 = correlation_matrix.loc[top_10_features, top_10_features]

plt.figure(figsize=(10, 8))
sns.heatmap(corr_top10, annot=True, cmap='vlag', fmt='.2f', linewidths=0.5)
plt.title(f'Top 10 Most Correlated Features with {target}')
#plt.savefig('V_most-correlated-features.pdf', bbox_inches='tight')
plt.show()
```

### Figure 6.3 &#x2601;		

```python
# mean per song
mean_df = df.groupby('file', as_index=False)[['Arousal', 'Valence']].mean()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=mean_df, x='Arousal', y='Valence', alpha=0.6, s=50, color="#C26F6D")

plt.xlabel('Arousal')
plt.ylabel('Valence')
plt.axhline(color='black', linestyle='--')
plt.axvline(color='black', linestyle='--')
#plt.savefig('scatter-plot.png')
plt.show()
```
