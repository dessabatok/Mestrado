### Packages &#x1F4DA;

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Figure 2.1 &#x1F509;	&#x1F50A;	

```python
t = np.linspace(0, 1, 1000)

# Primeira onda
freq1 = 5
amp1 = 1
wave1 = amp1 * np.sin(2 * np.pi * freq1 * t)

# Segunda onda
freq2 = 15
amp2 = 0.5
wave2 = amp2 * np.sin(2 * np.pi * freq2 * t)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# Primeiro gráfico
axs[0].plot(t, wave1, color='orange', label=f'Frequency: {freq1} Hz, Amplitude: {amp1}')
axs[0].set_title("Sound Wave")
axs[0].set_xlabel("Time (s)")
axs[0].set_ylabel("Amplitude")
axs[0].legend(loc='lower right')
axs[0].grid(True)
axs[0].set_ylim(-1.1, 1.1)

# Segundo gráfico
axs[1].plot(t, wave2, color='orange', label=f'Frequency: {freq2} Hz, Amplitude: {amp2}')
axs[1].set_title("Sound Wave with Higher Frequency and Lower Amplitude")
axs[1].set_xlabel("Time (s)")
axs[1].set_ylabel("Amplitude")
axs[1].legend(loc='lower right')
axs[1].grid(True)
axs[1].set_ylim(-1.1, 1.1)

plt.tight_layout()
plt.show()
```
