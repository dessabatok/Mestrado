### Packages &#x1F4DA;

```python
import numpy as np
import seaborn as sns
sns.set_theme()
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Figure 4.3 &#x1F514;	

```python
mu1, mu2 = 0, 7.5 
sigma1, sigma2 = 1, 1

x = np.linspace(-12, 12, 1000)

y1 = norm.pdf(x, mu1, sigma1)
y2 = norm.pdf(x, mu2, sigma2)

fig, axs = plt.subplots(2, 1, figsize=(8, 10))
axs[0].plot(x, y1, label=f' = {mu1},  = {sigma1}', color='purple')
axs[0].set_xlabel('x Value')
axs[0].set_ylabel('Probability Density')
axs[0].legend()
axs[0].grid(True)
axs[0].set_xlim(-3, 13)
axs[0].set_ylim(-0.01, 0.45)
axs[1].plot(x, y2, label=f' = {mu2},  = {sigma2}', color='green')
axs[1].set_xlabel('x Value')
axs[1].set_ylabel('Probability Density')
axs[1].legend()
axs[1].grid(True)
axs[1].set_xlim(-3, 13)
axs[1].set_ylim(-0.01, 0.45)

plt.legend()
plt.tight_layout()
plt.show()
```
